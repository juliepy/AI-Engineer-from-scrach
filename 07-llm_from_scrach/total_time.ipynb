```python
import os
import cv2
from datetime import timedelta

def get_total_duration(video_dir="video"):
    # 支持的视频格式
    video_extensions = ('.mp4', '.mkv', '.avi', '.mov', '.wmv', '.flv')
    total_seconds = 0
    
    # 遍历目录
    for root, dirs, files in os.walk(video_dir):
        for file in files:
            if file.lower().endswith(video_extensions):
                try:
                    video_path = os.path.join(root, file)
                    cap = cv2.VideoCapture(video_path)
                    
                    # 获取帧率和总帧数
                    fps = cap.get(cv2.CAP_PROP_FPS)
                    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
                    
                    # 计算时长（秒）
                    duration = frame_count / fps
                    
                    # 格式化单个视频时长
                    duration_formatted = str(timedelta(seconds=int(duration)))
                    # print(f"{video_path}: {duration_formatted}")
                    
                    total_seconds += duration
                    cap.release()
                except Exception as e:
                    print(f"Error processing {file}: {str(e)}")
    
    # 格式化总时长
    total_duration = str(timedelta(seconds=int(total_seconds)))
    print(f"\nTotal duration: {total_duration}")

# 使用前安装依赖
# pip install opencv-python

video_dir = r"D:\workspace\doc\智能实验室\coze-20260619\video"
# 运行函数
get_total_duration(video_dir)
```
