## Part 6 — 练习 B（进阶）：消息队列多 Agent Pipeline

### 题目

用 `queue.Queue` 模拟消息队列，实现 Pipeline 风格多 Agent 系统：

```
Input → [Q1] → RetrieverWorker → [Q2] → GeneratorWorker → [Q3] → VerifierWorker → Output
```

**要求：**
1. 实现三个 Worker 的 `process()` 方法
2. Worker 之间仅通过队列通信
3. 支持并发处理多条查询

````python
import queue
import threading
import time
from dataclasses import dataclass, field
from typing import Dict, Any


@dataclass
class Message:
    job_id: str
    stage: str
    payload: Dict[str, Any] = field(default_factory=dict)


class BaseWorker:
    def __init__(self, name: str, in_queue: queue.Queue, out_queue: queue.Queue):
        self.name = name
        self.in_queue = in_queue
        self.out_queue = out_queue

    def process(self, message: Message) -> Message:
        raise NotImplementedError

    def run(self):
        while True:
            try:
                msg = self.in_queue.get(timeout=2)
                if msg is None:
                    self.out_queue.put(None)
                    break
                print(f'  [{self.name}] 处理 {msg.job_id}')
                self.out_queue.put(self.process(msg))
                self.in_queue.task_done()
            except queue.Empty:
                break


class RetrieverWorker(BaseWorker):
    """TODO: 从知识库检索与 payload['query'] 相关的文档"""
    def process(self, message: Message) -> Message:
        time.sleep(0.05)
        # TODO: 使用 KNOWLEDGE_BASE 进行关键词检索
        return Message(message.job_id, 'retrieved',
                       {**message.payload, 'docs': ['占位文档']})


class GeneratorWorker(BaseWorker):
    """TODO: 调用 llm_sonnet 基于 payload['docs'] 生成答案"""
    def process(self, message: Message) -> Message:
        time.sleep(0.05)
        # TODO: 使用 LLM 生成答案
        return Message(message.job_id, 'generated',
                       {**message.payload, 'answer': '占位答案'})


class VerifierWorker(BaseWorker):
    """TODO: 验证 payload['answer'] 与 payload['docs'] 的一致性"""
    def process(self, message: Message) -> Message:
        time.sleep(0.05)
        # TODO: 调用 LLM 进行事实核查
        return Message(message.job_id, 'verified',
                       {**message.payload, 'passed': True})


# 运行 Pipeline
q1, q2, q3, q_out = queue.Queue(), queue.Queue(), queue.Queue(), queue.Queue()

workers = [
    RetrieverWorker('Retriever', q1, q2),
    GeneratorWorker('Generator', q2, q3),
    VerifierWorker('Verifier', q3, q_out),
]
threads = [threading.Thread(target=w.run, daemon=True) for w in workers]
for t in threads:
    t.start()

test_queries = [
    'AutoGen 的主要特性是什么？',
    'CrewAI 如何处理任务分配？',
    '什么是 Agentic RAG？',
]

print('📨 发送查询...')
for i, q in enumerate(test_queries):
    q1.put(Message(f'job_{i+1}', 'input', {'query': q}))
q1.put(None)  # 停止信号

results = []
print('\n📊 结果：')
while len(results) < len(test_queries):
    try:
        r = q_out.get(timeout=15)
        if r is None:
            break
        results.append(r)
        status = '✅' if r.payload.get('passed') else '❌'
        print(f'  {status} {r.job_id}: {r.payload.get("query", "")[:40]}')
    except queue.Empty:
        break

print(f'\n完成：{len(results)}/{len(test_queries)} 条')
````

## Part 7 — 练习 C（挑战）：Supervisor + HITL

### 题目

实现 **Supervisor** 逻辑：当 `pass_rate < 0.8` 时触发人工审查。

**要求：**
1. `Supervisor.should_review()` — 根据 `pass_rate` 判断
2. `Supervisor.run_hitl_review()` — CLI 交互（`input()`）
3. 集成到 Agentic RAG 流程中

````python
from collections import deque


class Supervisor:
    def __init__(self, window_size: int = 5, threshold: float = 0.8):
        self.window_size = window_size
        self.threshold = threshold
        self._recent: deque = deque(maxlen=window_size)
        self.review_queue: list = []
        self.decisions: list = []

    def record(self, passed: bool) -> None:
        self._recent.append(passed)

    @property
    def pass_rate(self) -> float:
        return sum(self._recent) / len(self._recent) if self._recent else 1.0

    def should_review(self) -> bool:
        # TODO: 当 pass_rate < threshold 时返回 True
        return False  # 替换为实际逻辑

    def add_to_queue(self, item: dict) -> None:
        self.review_queue.append(item)
        print(f'  [Supervisor] 加入审查队列 (pass_rate={self.pass_rate:.2f})')

    def run_hitl_review(self) -> None:
        """TODO: 遍历 review_queue，对每条用 input() 获取人工决策"""
        if not self.review_queue:
            print('[Supervisor] 审查队列为空')
            return
        print(f'\n🔍 人工审查（{len(self.review_queue)} 条待审）')
        for item in self.review_queue[:]:
            print(f'  查询：{item.get("query")}')
            print(f'  问题：{item.get("issues", [])}')
            # TODO: decision = input('批准(a)/拒绝(r): ').strip().lower()
            decision = 'a'  # 占位，替换为 input()
            self.decisions.append({'query': item['query'],
                                   'decision': 'approved' if decision == 'a' else 'rejected'})
            self.review_queue.remove(item)
        print(f'审查完成，{len(self.decisions)} 条决策已记录')


# 测试
sv = Supervisor(window_size=5, threshold=0.8)
for r in [True, True, False, False, False]:  # 3/5 = 60% < 80%
    sv.record(r)

print(f'pass_rate = {sv.pass_rate:.2f}')
print(f'should_review() = {sv.should_review()}  ← 实现后应为 True')
````
