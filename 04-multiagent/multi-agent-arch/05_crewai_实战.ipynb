## 三、CrewAI：定位与设计哲学

- CrewAI 官方文档把重点放在 agents、crews 与 flows
- 它更像“任务编排系统”，而不只是对话系统
- 关注的是任务分配、执行过程和可观测性
- 适合自动化流程、长期运行任务和可复用工作流
- 心智模型偏“交付流程”而不是“聊天过程”

> 对训练营学员来说，可以把它理解成“更接近生产的协作框架”。

---

## CrewAI 的 Agent / Task / Crew / Process

- Agent：有角色、目标、工具的执行者
- Task：带描述和期望输出的工作单元
- Crew：一组 Agents + Tasks 的协作组织
- Process：定义任务如何流动和被管理
- 这是一个很典型的“任务导向”抽象

课堂要点：

- 先想 task contract
- 再想 agent personality

---

## CrewAI 的 Sequential 与 Hierarchical

| 模式 | 核心特点 | 适合什么 |
| --- | --- | --- |
| Sequential | 任务按顺序执行 | 稳定线性流程 |
| Hierarchical | 由 manager 负责协调、委派和验收 | 更复杂的分工协作 |

官方文档提醒：

- Hierarchical 通常需要 `manager_llm` 或 `manager_agent`

---

## CrewAI 的 Tools、Knowledge 与 Memory

- Tools 负责调用外部能力
- Knowledge sources 负责给 Agent 提供知识背景
- Memory 负责保留连续性和上下文积累
- 结构化输出有助于 task 验收和系统对接
- 不要把所有知识都塞给所有 Agent

设计建议：

- 谁需要知识，谁拿知识
- 谁需要动作，谁拿工具

---

## CrewAI 的 Planning、Delegation 与资源意识

- Crew 可以在执行前加入 planning 阶段
- 委派通常通过 manager 或 task 分解体现
- 速率、预算和执行节奏需要有意识地控制
- 高成本生成步骤不要默认无限扩容
- 生产系统里往往要把 Crew 放入更大的调度环境

> 讲师提示：这里强调“成本和节奏是架构的一部分”。

---

## CrewAI 的 Retry、Health 与 Observability

- 官方文档提供 tracing、usage metrics 与 observability 能力
- 需要看见任务时间线、工具使用和失败位置
- Retry 必须与幂等设计一起考虑
- 健康检查不是“可选项”，而是长期运行前提
- 让每个 task 都能被解释、被复盘、被重试

一句话：

- Crew 跑起来以后，就进入了运维视角

---

## CrewAI 的优势与适用场景

- 适合长流程自动化和重复性交付
- 适合报告生成、运营流程、集成型自动化
- 适合想把任务所有权说清楚的团队
- 对接触发器、流程编排和观测时更自然
- 在“多步骤任务链”上表达力很强

典型场景：

- 研究报告
- 内容生产流水线
- 运营自动化

---

## CrewAI 的局限与不适合场景

- 如果只是一个很小的两步原型，可能显得偏重
- Crew 抽象并不会自动消除底层复杂度
- Agent 越多，任务边界越要严格，不然会混乱
- 真正的资源调度仍需要外部基础设施配合
- 不能把“有 Crew”误当成“已经生产可用”

注意：

- 任务编排清晰，不等于风险已经控制住

---

## CrewAI 工作流伪代码

```python
crew = Crew(
    agents=[researcher, writer, reviewer],
    tasks=[task1, task2, task3],
    process=Process.sequential,
)
result = crew.kickoff(inputs={"topic": topic})
```

- 真正生产中，常把 `kickoff()` 放进 queue、trigger 或 flow 后面
- 课堂重点：看 task contract 和 process 设计

---

## CrewAI 教学点

- CrewAI 更像任务流，而不是多轮聊天
- 提问方式要变成：谁负责什么、产出什么、失败怎么办
- 讲清 Agent 设计和 Process 设计是两件事
- 当流程可重复、可验收、可观测时，CrewAI 价值会更明显
- 不要只盯 prompt，要看任务链是否健康

> 互动：把“生成周报”拆成 3 个 Task，看看谁来负责最合适。

## Part 3 — CrewAI 实战

> **CrewAI** 提供高层级抽象，聚焦于「角色扮演」式的多 Agent 协作，适合长流程自动化任务。
> 通过设置 `OPENAI_API_BASE` 和 `OPENAI_API_KEY` 环境变量，CrewAI 会通过 LiteLLM 路由到 OpenRouter。

### 核心概念
- **`Agent`** — 有 `role`、`goal`、`backstory` 的智能体
- **`Task`** — 有明确 `description` 和 `expected_output` 的工作单元
- **`Crew`** — 将 Agents 和 Tasks 组织成协作团队
- **执行模式** — `Process.sequential`（默认）或 `Process.hierarchical`

````python
from crewai import Agent, Task, Crew, Process
from crewai import LLM

# 通过 CrewAI 的 LLM 类配置 OpenRouter
# CrewAI 使用 LiteLLM，支持 openrouter/ 前缀
crew_llm_sonnet = LLM(
    model=f'openrouter/{MODEL_SONNET}',
    api_key=OPENROUTER_API_KEY,
    base_url=OPENROUTER_BASE_URL,
)

crew_llm_haiku = LLM(
    model=f'openrouter/{MODEL_HAIKU}',
    api_key=OPENROUTER_API_KEY,
    base_url=OPENROUTER_BASE_URL,
)

print('✅ CrewAI LLM 配置完成（通过 OpenRouter）')
print(f'   Sonnet: openrouter/{MODEL_SONNET}')
print(f'   Haiku : openrouter/{MODEL_HAIKU}')
````

### 示例 3A — 顺序流水线 Crew：内容生成

三个 Agent 依次执行：研究 → 写作 → 审校

```
[research_task] ──► [writing_task] ──► [editing_task]
     │                    │                  │
 Researcher            Writer             Editor
```

````python
# 定义 Agents
crew_researcher = Agent(
    role='技术研究员',
    goal='深入研究指定主题，收集关键事实、数据和见解',
    backstory=(
        '你是一位经验丰富的技术研究员，擅长快速定位信息要点。'
        '你的研究结果简洁准确，为后续写作提供坚实基础。'
    ),
    llm=crew_llm_haiku,
    allow_delegation=False,
    verbose=True,
)

crew_writer = Agent(
    role='内容写作者',
    goal='将研究成果转化为引人入胜的 3 段式文章',
    backstory=(
        '你是一位专业内容写作者，擅长将复杂技术信息转化为易读内容。'
        '你的文章结构清晰，语言生动，能够吸引技术和非技术读者。'
    ),
    llm=crew_llm_sonnet,
    allow_delegation=False,
    verbose=True,
)

crew_editor = Agent(
    role='内容审校者',
    goal='确保文章质量，消除错误，提升可读性',
    backstory=(
        '你是一位严格的内容审校者，有多年编辑经验。'
        '你不仅纠正语法错误，还关注逻辑连贯性和整体质量。'
    ),
    llm=crew_llm_haiku,
    allow_delegation=False,
    verbose=True,
)

# 定义 Tasks
research_task = Task(
    description=(
        '研究主题：{topic}\n\n'
        '请提供：\n'
        '1. 主题的 3-5 个核心知识点\n'
        '2. 2-3 个实际应用案例\n'
        '3. 1-2 个未来趋势预测\n'
        '总字数控制在 300 字以内。'
    ),
    expected_output='结构化的研究摘要，包含核心知识点、应用案例和趋势预测',
    agent=crew_researcher,
)

writing_task = Task(
    description=(
        '基于研究员提供的研究摘要，撰写一篇关于 {topic} 的文章。\n\n'
        '要求：\n'
        '- 3 个段落：引言（吸引读者）、主体（核心内容）、结论（展望未来）\n'
        '- 每段 100-150 字\n'
        '- 语言流畅，面向有技术背景的读者'
    ),
    expected_output='完整的 3 段式文章，共约 300-450 字',
    agent=crew_writer,
)

editing_task = Task(
    description=(
        '审校写作者提交的文章。\n\n'
        '检查内容：\n'
        '1. 语法和表达是否准确\n'
        '2. 段落结构是否清晰\n'
        '3. 与研究摘要是否一致\n\n'
        '输出经过修改和完善的最终版本文章。'
    ),
    expected_output='经过审校的最终文章，附带简短的修改说明（最多 3 条）',
    agent=crew_editor,
)

# 组建 Crew
content_crew = Crew(
    agents=[crew_researcher, crew_writer, crew_editor],
    tasks=[research_task, writing_task, editing_task],
    process=Process.sequential,
    verbose=True,
)

print('✅ Content Crew 构建完成（顺序模式）')
````

````python
# 执行 Crew
result = content_crew.kickoff(inputs={'topic': '多 Agent AI 系统在企业中的落地实践'})

print('\n' + '=' * 60)
print('📄 最终输出：')
print('=' * 60)
print(result)
````

### 示例 3B — 层级制 Crew（Hierarchical Process）

在层级制中，CrewAI 自动创建 **Manager Agent** 负责任务分配和协调，其他 Agent 作为 Worker。

```
        Manager (LLM)
       /             \
  TechAnalyst    RiskAnalyst
```

````python
tech_analyst = Agent(
    role='技术分析师',
    goal='评估技术方案的可行性和技术细节',
    backstory='你是资深技术分析师，专注于技术可行性评估和架构分析。',
    llm=crew_llm_haiku,
    allow_delegation=False,
    verbose=True,
)

risk_analyst = Agent(
    role='风险评估师',
    goal='识别潜在风险，制定缓解策略',
    backstory='你专注于识别技术和业务风险，并提供实用的风险缓解建议。',
    llm=crew_llm_haiku,
    allow_delegation=False,
    verbose=True,
)

review_task = Task(
    description=(
        '对以下技术方案进行全面评审：\n\n'
        '{proposal}\n\n'
        '要求：\n'
        '1. 技术可行性分析\n'
        '2. 风险识别与缓解建议\n'
        '3. 综合评审结论'
    ),
    expected_output='包含技术分析、风险评估和综合结论的评审报告',
    agent=tech_analyst,
)

hierarchical_crew = Crew(
    agents=[tech_analyst, risk_analyst],
    tasks=[review_task],
    process=Process.hierarchical,
    manager_llm=crew_llm_sonnet,
    verbose=True,
)

proposal = """
方案：在现有电商平台引入多 Agent AI 客服系统
- 部署 5 个专业 Agent（订单查询、退款处理、产品推荐、投诉处理、人工转接）
- 预计处理 80% 的客服请求，响应时间 < 3 秒
- 第一期预算：50 万元，3 个月交付
"""

result_hierarchical = hierarchical_crew.kickoff(inputs={'proposal': proposal})
print('\n📋 评审报告：')
print(result_hierarchical)
````

### 示例 3C — 带工具的 Crew（网络搜索）

````python
if not TAVILY_API_KEY:
    print('⚠️  跳过示例 3C：TAVILY_API_KEY 未设置')
    print('   请在 .env 文件中添加 TAVILY_API_KEY 后重新运行')
else:
    from crewai_tools import TavilySearchTool

    search_tool = TavilySearchTool(api_key=TAVILY_API_KEY)

    web_researcher = Agent(
        role='网络研究员',
        goal='通过网络搜索获取最新信息',
        backstory='你擅长使用搜索工具获取准确、最新的技术信息。',
        llm=crew_llm_haiku,
        tools=[search_tool],
        allow_delegation=False,
        verbose=True,
    )

    search_task = Task(
        description='搜索 2025 年多 Agent AI 框架的最新进展，重点关注 AutoGen 和 CrewAI 的新特性。',
        expected_output='3-5 条关于 AutoGen 和 CrewAI 最新进展的摘要，附带来源',
        tools=[search_tool],
        agent=web_researcher,
    )

    search_crew = Crew(
        agents=[web_researcher],
        tasks=[search_task],
        verbose=True,
    )

    search_result = search_crew.kickoff()
    print('\n🔍 搜索结果：')
    print(search_result)
````
